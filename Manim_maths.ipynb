```python
import matplotlib.pyplot as plt
import numpy as np
import colorsys

# Setup figure
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_facecolor("black")
ax.axis("off")

x, y = 0, 0
angle = 0
h = 0.5

# Store points
points_x = []
points_y = []
colors = []

# Simulate turtle-like drawing
for i in range(360):
    r, g, b = colorsys.hsv_to_rgb(h, 1, 1)
    color = (r, g, b)
    h += 0.005

    for j in range(5):
        x += np.cos(np.radians(angle)) * 10
        y += np.sin(np.radians(angle)) * 10
        points_x.append(x)
        points_y.append(y)
        colors.append(color)
        angle += 5

    angle += 65
    for _ in range(30):  # simulate circle-like curve
        x += np.cos(np.radians(angle)) * 10
        y += np.sin(np.radians(angle)) * 10
        points_x.append(x)
        points_y.append(y)
        colors.append(color)
        angle += 12
    angle += 65

# Plot the path
ax.scatter(points_x, points_y, c=colors, s=2)
plt.show()
```

```python
%%writefile triangle_concepts.py
from manim import *

# Color palette
TRIANGLE_COLOR = BLUE
HIGHLIGHT_COLOR = YELLOW
LABEL_COLOR = WHITE

# Scene 1: Types of Triangles
class TypesOfTriangles(Scene):
    def construct(self):
        title = Text("Types of Triangles", font_size=40).to_edge(UP)
        mascot = Text("🐢", font_size=40).next_to(title, LEFT)

        eq_triangle = Polygon([0,0,0], [2,0,0], [1,1.73,0], color=TRIANGLE_COLOR)
        eq_label = Text("Equilateral", font_size=30).next_to(eq_triangle, DOWN)

        iso_triangle = Polygon([4,0,0], [6,0,0], [5,1.5,0], color=TRIANGLE_COLOR)
        iso_label = Text("Isosceles", font_size=30).next_to(iso_triangle, DOWN)

        sc_triangle = Polygon([8,0,0], [10,0,0], [9.3,1.2,0], color=TRIANGLE_COLOR)
        sc_label = Text("Scalene", font_size=30).next_to(sc_triangle, DOWN)

        self.play(Write(title), FadeIn(mascot))
        self.play(DrawBorderThenFill(eq_triangle), Write(eq_label))
        self.play(DrawBorderThenFill(iso_triangle), Write(iso_label))
        self.play(DrawBorderThenFill(sc_triangle), Write(sc_label))
        self.wait(2)
        self.play(*[FadeOut(mob) for mob in self.mobjects])

# Scene 2: Angle Sum Property
class AngleSumProperty(Scene):
    def construct(self):
        title = Text("Angle Sum Property", font_size=40).to_edge(UP)
        mascot = Text("🦉", font_size=40).next_to(title, LEFT)

        triangle = Polygon([-2,0,0], [2,0,0], [0,2,0], color=TRIANGLE_COLOR)
        A = triangle.get_vertices()[0]
        B = triangle.get_vertices()[1]
        C = triangle.get_vertices()[2]
        angle_labels = [MathTex(r"\angle A"), MathTex(r"\angle B"), MathTex(r"\angle C")]
        angle_labels[0].next_to(A, LEFT + DOWN)
        angle_labels[1].next_to(B, RIGHT + DOWN)
        angle_labels[2].next_to(C, UP)

        sum_text = MathTex(r"\angle A + \angle B + \angle C = 180^\circ").to_edge(DOWN)

        self.play(Write(title), FadeIn(mascot))
        self.play(Create(triangle))
        self.play(*[Write(label) for label in angle_labels])
        self.wait(1)
        self.play(Write(sum_text))
        self.wait(2)
        self.play(*[FadeOut(mob) for mob in self.mobjects])

# Scene 3: Pythagoras Theorem
class PythagorasTheorem(Scene):
    def construct(self):
        title = Text("Pythagoras Theorem", font_size=40).to_edge(UP)
        mascot = Text("🐇", font_size=40).next_to(title, LEFT)

        A = [0, 0, 0]
        B = [3, 0, 0]
        C = [0, 4, 0]
        triangle = Polygon(A, B, C, color=TRIANGLE_COLOR)

        square_a = Square(3, color=GREEN).move_to([1.5, -1.5, 0])
        square_b = Square(4, color=ORANGE).move_to([-2, 2, 0])
        square_c = Square(np.sqrt(25), color=RED).move_to([2.5, 2, 0]).rotate(-PI/2)

        theorem = MathTex("a^2 + b^2 = c^2").to_edge(DOWN)

        self.play(Write(title), FadeIn(mascot))
        self.play(Create(triangle))
        self.wait(0.5)
        self.play(FadeIn(square_a), FadeIn(square_b))
        self.wait(1)
        self.play(FadeIn(square_c))
        self.wait(1)
        self.play(Write(theorem))
        self.wait(2)
        self.play(*[FadeOut(mob) for mob in self.mobjects])

# Scene 4: Exterior Angle Property
class ExteriorAngleProperty(Scene):
    def construct(self):
        title = Text("Exterior Angle Property", font_size=40).to_edge(UP)
        mascot = Text("🐱", font_size=40).next_to(title, LEFT)

        triangle = Polygon([-2,0,0], [2,0,0], [1,2,0], color=TRIANGLE_COLOR)
        extended_line = Line([-2,0,0], [-3,0,0], color=GRAY)
        exterior_angle = Arc(radius=0.8, start_angle=0, angle=PI/4, arc_center=[-2,0,0], color=RED)
        angle_text = MathTex(r"\text{Exterior Angle} = \angle B + \angle C").to_edge(DOWN)

        self.play(Write(title), FadeIn(mascot))
        self.play(Create(triangle), Create(extended_line))
        self.play(Create(exterior_angle))
        self.wait(1)
        self.play(Write(angle_text))
        self.wait(2)
        self.play(*[FadeOut(mob) for mob in self.mobjects])

# Scene 5: Area of Triangle
class AreaOfTriangle(Scene):
    def construct(self):
        title = Text("Area of a Triangle", font_size=40).to_edge(UP)
        mascot = Text("🦊", font_size=40).next_to(title, LEFT)

        base = Line([-2, 0, 0], [2, 0, 0], color=BLUE)
        height = Line([0, 0, 0], [0, 3, 0], color=GREEN)
        triangle = Polygon([-2, 0, 0], [2, 0, 0], [0, 3, 0], color=TRIANGLE_COLOR)

        base_label = MathTex("b").next_to(base, DOWN)
        height_label = MathTex("h").next_to(height, LEFT)
        formula = MathTex(r"\text{Area} = \frac{1}{2} \times b \times h").to_edge(DOWN)

        self.play(Write(title), FadeIn(mascot))
        self.play(Create(base), Create(height))
        self.play(Create(triangle))
        self.play(Write(base_label), Write(height_label))
        self.wait(1)
        self.play(Write(formula))
        self.wait(2)
        self.play(*[FadeOut(mob) for mob in self.mobjects])

# Master Scene
class AllTriangleConcepts(Scene):
    def construct(self):
        self.add(Text("Let's Learn About Triangles! 🧠", font_size=48).to_edge(UP))
        self.wait(1)
        self.play(FadeOut(*self.mobjects))
        for scene_class in [TypesOfTriangles, AngleSumProperty, PythagorasTheorem, ExteriorAngleProperty, AreaOfTriangle]:
            scene = scene_class()
            scene.construct()
```

```python
%%writefile exterior_angle.py
from manim import *

class ExteriorAngle(Scene):
    def construct(self):
        # 1. Setup triangle points
        A = LEFT * 2 + DOWN * 1
        B = RIGHT * 2 + DOWN * 1
        C = UP * 2

        # 2. Draw triangle & labels
        tri = Polygon(A, B, C, color=BLUE)
        self.play(Create(tri))
        labels = VGroup(
            MathTex("A").next_to(A, LEFT + DOWN),
            MathTex("B").next_to(B, RIGHT + DOWN),
            MathTex("C").next_to(C, UP)
        )
        self.play(Write(labels))

        # 3. Extend side AB beyond B to draw exterior angle at B
        extension = Line(B, B + (B - A) * 0.8, color=GRAY)
        self.play(Create(extension))

        # 4. Draw interior angles A, B, C
        angle_A = Angle(Line(A, C), Line(A, B), radius=0.5, color=YELLOW)
        angle_B = Angle(Line(B, A), Line(B, C), radius=0.5, color=GREEN)
        angle_C = Angle(Line(C, B), Line(C, A), radius=0.5, color=ORANGE)
        self.play(Create(angle_A), Create(angle_B), Create(angle_C))

        # 5. Label interior angles
        self.play(
            Write(MathTex(r"\alpha").move_to(angle_A.get_arc_center() + LEFT * 0.2)),
            Write(MathTex(r"\beta").move_to(angle_B.get_arc_center() + DOWN * 0.2)),
            Write(MathTex(r"\gamma").move_to(angle_C.get_arc_center() + UP * 0.2)),
        )

        # 6. Draw and label exterior angle at B (supplementary to interior at B)
        ext_angle = Angle(Line(B, C), extension, radius=0.8, other_angle=True, color=RED)
        self.play(Create(ext_angle))
        self.play(Write(MathTex(r"\angle_{\rm ext}").next_to(ext_angle, RIGHT * 0.3 + DOWN * 0.1)))

        # 7. Display the relationship equation
        eq = MathTex(r"\angle_{\rm ext} = \alpha + \beta").to_edge(DOWN)
        self.wait(1)
        self.play(Write(eq))

        self.wait(2)
```
